```python
from operator import indexOf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error
from sklearn import metrics

pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_colwidth',1000)

df_LIMPIO=pd.read_csv('data_NEW/df_LIMPIO.csv', index_col=0)
df_LIMPIO
```

```python
# ELIMINAR OUTLIERS LOQUÍSIMOS ES VERY VERY IMPORTANTE
import scipy.stats as stats
from scipy.stats import mstats


df_LIMPIO['price_per_year'] = stats.mstats.winsorize(df_LIMPIO['price_per_year'], limits=[None, 0.4])
# df_LIMPIO['price_per_year'] = np.log1p(df_LIMPIO['price_per_year'])
```

```python
X=df_LIMPIO.drop(columns=[
                        ## VARIABLES TARGET
                        'price',
                        'price_per_month',
                        'price_per_year',

                        ## VARIABLES PREDICTORAS ÚTILES
                        # 'latitude',
                        # 'longitude',
                        # 'accommodates',
                        # 'bedrooms',
                        'beds',
                        'amenities_len',
                        # 'room_type_code',
                        'eur_distrito',
                        'Ranking',
                        'dist_p_sol',
                        # 'bathrooms_num',
                        # 'bathrooms_priv_or_shar',
                        'facturas_inc',
                        'pet',
                        'airport',
                        'lux',
                        'breakfast',
                        'wifi',
                        'pool',
                        # 'air_cond',
                        # 'elevator',
                        'balcony',
                        'workspace',
                        'nespresso',
                        'single_level',

                        ## DATOS SOBRE HOST Y REVIEWS
                        'reviews_per_month',
                        'review_scores_rating',
                        'host_since'
                        # 'total_viviendas_host',
 ])
        
y=df_LIMPIO['price_per_year']

features = list(X)

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state=42)
```

```python
# prueba = X_test
nueva_fila = {'latitude':40.403002,
            'longitude': -3.690574,
            'accommodates': 3,
            'bedrooms': 2,
            'total_viviendas_host': 2,
            'room_type_code': 100,
            'bathrooms_num':1,
            'bathrooms_priv_or_shar': 1,
            'air_cond': 1,
            'elevator': 1
            }

prueba= X_test.append(nueva_fila, ignore_index=True)

streamlit_test = prueba[-1:]
streamlit_test
```

```python
streamlit_test.to_csv('data_NEW/streamlit_test.csv')
```

```python
modelo_streamlit = RandomForestRegressor(
            n_estimators = 450,
            criterion    = 'squared_error',
            max_depth    = 50,
            max_features = 4,
            oob_score    = False,
            n_jobs       = -1,
            random_state = 123
            )
            
modelo_streamlit.fit(X_train, y_train)

predictions_rf = modelo_streamlit.predict(X_test)

print('MAE:', mean_absolute_error(y_test, predictions_rf))
# print('MSE:', mean_squared_error(y_test, predictions_rf))
# print('RMSE:', np.sqrt(mean_squared_error(y_test, predictions_rf)))

print(modelo_streamlit.score(X_train, y_train))
print(modelo_streamlit.score(X_test, y_test))
```
